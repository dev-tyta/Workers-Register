```python
import cv2
import face_recognition
```

```python
image = face_recognition.load_image_file("../data/player_data/hcantoni.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
face = face_recognition.face_locations(image)[0]
image_copy = image.copy()
```

```python
cv2.rectangle(image, (face[3], face[0]), (face[2], face[1]), (0, 0, 255))
cv2.imshow("Face Recognition", image)
cv2.imshow("Real Face", image_copy)
cv2.waitKey(0)
```
